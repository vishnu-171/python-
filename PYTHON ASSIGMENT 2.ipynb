```python
filename = "count_lines_unix.txt"
try:
    with open(filename, 'r') as infile:
        total_lines = sum(1 for _ in infile)
    print(f"Total number of lines in '{filename}': {total_lines}")
except FileNotFoundError:
    print("Error: File not found. Please check the filename and try again.")

```

```python
file_name = "display_last_before_line.txt"
try:
    with open(file_name, 'r') as f:
        all_lines = f.readlines()
        if len(all_lines) >= 2:
            print("Second-last line content:\n", all_lines[-2].strip())
        else:
            print("File has less than two lines.")
except FileNotFoundError:
    print("Error: File not found.")
```

```python
src = "source_file.txt"
dest = "copied_file.txt"
try:
    with open(src, 'r') as infile, open(dest, 'w') as outfile:
        for line in infile:
            outfile.write(line)
    print(f"File '{src}' has been copied to '{dest}' successfully.")
except FileNotFoundError:
    print("Error: Source file not found.")
```

```python
fname = "count_word_occurrence.txt"
word_to_find = "python"
try:
    with open(fname, 'r') as f:
        content = f.read().lower()
        word_count = content.split().count(word_to_find.lower())
        print(f"The word '{word_to_find}' appears {word_count} times in the file.")
except FileNotFoundError:
    print("Error: File not found.")
```

```python
filename = "divide_numbers.txt"
try:
    with open(filename, 'r') as f:
        for line in f:
            a, b = map(int, line.split())
            print(a / b)
except FileNotFoundError:
    print("Error: File not found.")
except ZeroDivisionError:
    print("Error: Division by zero occurred.")
```

```python
filename = "exception_input.txt"
try:
    with open(filename, 'r') as f:
        total = 0
        for line in f:
            try:
                total += int(line.strip())
            except ValueError:
                continue
        print("Sum of all valid integers:", total)
except FileNotFoundError:
    print("Error: File not found.")
```

```python
filename = "custom_exception.txt"
try:
    with open(filename, 'r') as f:
        for line in f:
            num = int(line.strip())
            if num < 0:
                raise ValueError("Negative number found!")
            print(num)
except FileNotFoundError:
    print("Error: File not found.")
except ValueError as e:
    print(e)
```

```python
import math

numbers_file = "square_roots.txt"
try:
    with open(numbers_file, 'r') as f:
        for line in f:
            num = float(line.strip())
            if num < 0:
                raise ValueError("Cannot find square root of negative number.")
            print("Square root:", math.sqrt(num))
except FileNotFoundError:
    print("Error: File not found.")
except ValueError as err:
    print(err)
```

```python
import re
text = "My phone number is 9876543210."
pattern = r"\d{10}"
match = re.search(pattern, text)
if match:
    print("Phone number found:", match.group())
else:
    print("No phone number found.")
```

```python
import re
sentence = "The rain in Spain stays mainly in the plain."
pattern = r"\bin\b"
matches = re.findall(pattern, sentence)
print("Occurrences of 'in':", len(matches))
```

```python
import re
email_text = "Contact us at helpdesk@domain.com or admin@site.org"
pattern = r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}"
emails = re.findall(pattern, email_text)
print("Emails found:", emails)
```

```python
import re
log_data = "2025-10-29 10:15:23 ERROR File not found"
pattern = r"\d{4}-\d{2}-\d{2}"
date_match = re.search(pattern, log_data)
if date_match:
    print("Date found:", date_match.group())
else:
    print("No date found.")
```
